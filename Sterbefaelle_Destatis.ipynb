```python
import pandas as pd
```

```python
# Sterbefälle Bundesländer nach Kalenderwochen - Jahresangaben in Excel Tabellenblatt beachten
sterbefaelle = pd.read_excel("https://www.destatis.de/DE/Themen/Gesellschaft-Umwelt/Bevoelkerung/Sterbefaelle-Lebenserwartung/Tabellen/sonderauswertung-sterbefaelle.xlsx?__blob=publicationFile","BL_2016_2021_KW_AG_Ins",skiprows=8)
```

```python
# Sterbefälle Deutschland nach Kalenderwochen - Jahresangaben in Excel Tabellenblatt beachten
sterbefaelleD = pd.read_excel("https://www.destatis.de/DE/Themen/Gesellschaft-Umwelt/Bevoelkerung/Sterbefaelle-Lebenserwartung/Tabellen/sonderauswertung-sterbefaelle.xlsx?__blob=publicationFile","D_2016_2022_KW_AG_Ins",skiprows=8)
```

```python
sterbefaelleInsgesamt = sterbefaelle[sterbefaelle["unter … Jahren"]=="Insgesamt"]
```

```python
sterbefaelleDInsgesamt = sterbefaelleD[sterbefaelleD["unter … Jahren"]=="Insgesamt"]
```

```python
sterbefaelleInsg = sterbefaelleInsgesamt.drop(columns=["unter … Jahren","Nr."]).replace({"Baden-Württemberg" : "BW","Bayern" : "BY","Berlin" : "BE","Brandenburg" : "BB","Bremen" : "HB","Hamburg" : "HH","Hessen" : "HE","Mecklenburg-Vorpommern" : "MV","Niedersachsen" : "NI","Nordrhein-Westfalen" : "NW","Rheinland-Pfalz" : "RP","Saarland" : "SL","Sachsen" : "SN","Sachsen-Anhalt" : "ST","Schleswig-Holstein" : "SH","Thüringen" : "TH"})
```

```python
sterbefaelleInsg
```

```python
sterbefaelleDInsg = sterbefaelleDInsgesamt.drop(columns=["Nr."]).rename(columns={"unter … Jahren": "Unnamed: 2"}).replace({"Insgesamt" : "DE"})
```

```python
sterbefaelleDInsg
```

```python
sterbefaelleInsg=sterbefaelleInsg.append(sterbefaelleDInsg)
```

```python
sterbefaelleInsg["weekNr"] = sterbefaelleInsg["Unnamed: 2"] + sterbefaelleInsg["Unnamed: 1"].astype(str)
```

```python
sterbefaelleInsg
```

```python
# transpose output for d3js usage, create useful column names
output = sterbefaelleInsg.drop(columns=["Unnamed: 2","Unnamed: 1"]).set_index("weekNr").T
```

```python
output
```

```python
# save as csv for use in a d3js graph, you may want to archive this by date
output.to_csv("overall_mortality_weekly_germany_states.csv",index_label="weekNr",float_format='%.0f')
```
